```python
#Pregunta 3
def generar_condicion_edad(fila):
    edad = fila['Age']
    if (edad >= 18) :
        return 'Si'
    elif (edad < 18 ):
        return 'No'
    else:
        return 'No definido'

df["edad_mayor_menor"] = df.apply(generar_condicion_edad, axis=1)
#puse el drop porque cree dos columas con la misma condicion si es mayor o menor
df.drop(['condicion_edad'], axis=1)
```
