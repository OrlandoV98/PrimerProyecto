```python
# CONSTANTES
OWNER = 'CSSEGISandData'
REPO = 'COVID-19'
PATH = 'csse_covid_19_data/csse_covid_19_daily_reports'
URL = f'https://api.github.com/repos/{OWNER}/{REPO}/contents/{PATH}'
```

```python
URL
```

```python
import requests
import tqdm
download_urls = []
response = requests.get(URL)
for data in tqdm.tqdm(response.json()):
    if data['name'].endswith('.csv'):
        download_urls.append(data['download_url'])
```

```python
labels = ['Province_State', 'Country_Region', 'Last_Update', 'Confirmed', 'Deaths', 'Recovered']
```

```python
relabel = {
    # 'Last Update': 'Last_Update',
    'Country/Region': 'Country_Region',
    'Lat': 'Latitude',
    'Long_': 'Longitude',
    'Province/State': 'Province_State',
}


def factor_dataframe(dat, filename):
    """ Refactorizar el dataframe para ser montado a base datos SQL 
    como un DataFrame de pandas 
    """
    # renombrar etiquetas
    for label in dat:
        if label in relabel:
            dat = dat.rename(columns = {label: relabel[label]})
    
    # retornar un dataframe con estos parametros
    labels = ['Province_State', 'Country_Region', 'Last_Update', 'Confirmed', 'Deaths', 'Recovered']
    # nombre del archivo es el datetime
    if 'Last_Update' not in dat:
        dat['Last_Update'] = pd.to_datetime(filename)
    dat['Date'] = pd.to_datetime(filename)
    # remplazar columnas vacias con nan
    for label in labels:
        if label not in dat:
            dat[label] = np.nan

    return dat[labels]

def upload_to_sql(filenames, db_name, debug=False): 
    conn =  create_engine("mssql+pyodbc://" + name + ":" + pword + "@" + host +"/" + database + "?driver=ODBC+Driver+17+for+SQL+Server")

    
    if debug:
        print("Uploading into database")
    for i, file_path in tqdm.tqdm(list(enumerate(filenames))):
        
        dat = pd.read_csv(file_path)

        # renombrar etiquetas
        filename = os.path.basename(file_path).split('.')[0]
        dat = factor_dataframe(dat, filename)

        # escribir datos a sqlite
        if i == 0: # si es la primera entrada y el nombre de la tabla ya existe, entonces remplazar
            dat.to_sql(db_name, con=conn, index = False, if_exists='replace')
        else: # de lo contrario añadir a la tabla existente
            dat.to_sql(db_name, con=conn, index = False, if_exists='append')


# cargar datos a sqlite
upload_to_sql(download_urls, 'example', debug=True)
```

```python
#Data de ventas
# En este laboratorio crearemos un ETL que genere un reporte mensual de las ventas y algunas vistas transformadas sobre el dataset
path = r"C:\Users\LUIGGI\Desktop\Curso_ETL_Python\ETL DATASET - 2\Dataset_ ventas"

df = pd.read_excel(path +"\\Sample - Superstore.xlsx")
```

```python
#examinemos el dataset
df.head(5)
```

```python
#veamos la calidad de la data del dataframe
df.isna().sum()
```

```python
#si el profit no tiene nulos se podria estimar las ventas a partir de un % de la ganancia
df['Profit'].sum()/df['Sales'].sum()
```

```python
# cañculemos para los casos donde no se tengan el valor de la venta , lo remplazamos el valor profit/0.12

df['Sales'] = np.where(df['Sales'].isna(), df['Profit']/0.12, df['Sales'])
```

```python
#Sobre los datos que son categoricos se realizara un cambio del nulo por el estring "sin informacion"

df['Product ID']  = df['Product ID'].fillna("Sin Info")     
df['Category']    = df['Category'].fillna("Sin Info")  
df['Sub-Category']   = df['Sub-Category'].fillna("Sin Info")            
df['Product Name']   = df['Product Name'].fillna("Sin Info")         
```

```python
#veamos si aun quedan valores nulos
df.isna().sum()
```

```python
#construyamos el reporte mensualizado por mes y año y producto

df['Mes'] = df['Order Date'].dt.month
df['Año'] = df['Order Date'].dt.year
```

```python
#construimos el reporte de año y mes
reporte_mes_año = df.groupby(['Mes','Año']).agg({"Sales":"sum","Profit":"sum"}).reset_index()
```

```python
#Miremos el reporte
reporte_mes_año.head(5)
```

```python
# Creemos algunas variables
# calculemos el casto de fabricación como el 50% de las ventas
# Calculamos el IGV como el 18% de las ventas
reporte_mes_año["Costo_ventas"] = reporte_mes_año["Sales"] * 0.5
reporte_mes_año["IGV"] = reporte_mes_año["Sales"] * 0.18
```

```python
#Miremos el reporte
reporte_mes_año.head(5)
```

```python
#Generemos el reporte por regiones y sus manager
```

```python
#Maestro de regiones y managet
df_region = pd.read_excel(path +"\\Sample - Superstore.xlsx", sheet_name = "People")
```

```python
#Muestra de la nueva base de datos
df_region
```

```python
#realizaremos un merge
df_union = df.merge(df_region, on="Region", how = "left")
```

```python
# Reporte por gerente zonales
reporte_gerente_zonal = df_union.groupby(["Person"]).agg({"Sales":"sum","Profit":"sum"}).reset_index()
```

```python
# Creemos algunas variables
# calculemos el casto de fabricación como el 50% de las ventas
# Calculamos el IGV como el 18% de las ventas
reporte_gerente_zonal["Costo_ventas"] = reporte_gerente_zonal["Sales"] * 0.5
reporte_gerente_zonal["IGV"] = reporte_gerente_zonal["Sales"] * 0.18
```

```python
#Calculemos el ranking
reporte_gerente_zonal["Rank"] = reporte_gerente_zonal["Sales"].rank(ascending=False)
```

```python
reporte_gerente_zonal.head()
```

```python
#Finalmente realizaremos la carga de ambos reportes
pword = "Luiggi12."
name ="DB"
host = 'DESKTOP-LPN6KNV\MSQL'
database ='ETL'
#Carga desde SQL
import pyodbc
cnxn = pyodbc.connect(driver='{SQL Server Native Client 11.0}', 
                      host= host, 
                      database= database,
                      trusted_connection='yes',
                      user=name, password=pword)

#cur=cnxn.cursor()
#cur.execute("TRUNCATE TABLE CONFIRMADOS")  
#cur.commit()
cnxn.close()
engine = create_engine("mssql+pyodbc://" + name + ":" + pword + "@" + host +"/" + database + "?driver=ODBC+Driver+17+for+SQL+Server")

reporte_mes_año.to_sql('REPORTE_MENSUAL', con = engine, if_exists = 'append', chunksize = 10000)
reporte_gerente_zonal.to_sql('REPORTE_GERENTES', con = engine, if_exists = 'append', chunksize = 10000)
```
